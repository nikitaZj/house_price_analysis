```python
# 时间：2022/1/1 21:57
import requests
from lxml import etree
import time
import csv
class Lianjiaspider(object):
    def __init__(self):
#         self.url='https://sh.lianjia.com/ershoufang/pg{}/'
        self.url = 'https://sh.lianjia.com/ershoufang/pg{}/'
#         self.henders={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'}
        
    def get_page(self,url):
        time.sleep(3)
        self.henders={'User-Agent':getUserAgent()}
        respones=requests.get(url=url,headers=self.henders)
        respones.encoding='utf-8'
        html=respones.text
#         print(html)
        return self.parse_page(html)
        
    def parse_page(self,html):
        paser_html=etree.HTML(html)
        parse_page_result = {}
#         li_list = paser_html.xpath('//ul[@class="sellListContent"]/li[@class="clear LOGCLICKDATA"]/div[1]')
        li_list = paser_html.xpath('//*[@id="content"]/div[1]/ul/li/div[1]')
        #//li[@class="clear LOGCLICKDATA"]
        print('li_list:',li_list)
        
        for i in range(len(li_list)):
            li = li_list[i]
            house_dict = {}
            # 名称
            house_dict['house_name'] = li.xpath('.//div[@class="flood"]/div/a[1]/text()')#.strip()
            # 总价
            house_dict['total_price']=li.xpath('.//div[@class="totalPrice totalPrice2"]/span[1]/text()')
             # 总价单位
            house_dict['total_price_desc']=li.xpath('.//div[@class="totalPrice totalPrice2"]/i[2]/text()')
            # 单价
            house_dict['unit_price']=li.xpath('.//div[@class="unitPrice"]/span[1]/text()')#[0].strip()
           # 房屋其他信息 例：3室2厅 | 168.46平米 | 东南 | 精装 | 低楼层(共11层) | 1999年建 | 板楼
            house_dict['info']=li.xpath('.//div[@class="houseInfo"]/text()')
            # 房屋地址
            house_dict['location1']=li.xpath('.//div[@class="flood"]/div/a[2]/text()')

            print(house_dict)
            parse_page_result[i] = house_dict
        return parse_page_result
            
    def main(self):
        self.main_result = {}
        for pg in range(1,101):
            print('page:',pg)
            url=self.url.format(pg)
            self.main_result['pg'+str(pg)] = self.get_page(url)

if __name__ == "__main__":
    start=time.time()
    lianjia=Lianjiaspider()
    lianjia.main()
    end=time.time()
    print('执行时间为：%0.2f'%(end - start))


```

```python
# 保存
def pkl_save(filename, file):
    #将file保存到filename中
    output = open(filename, 'wb')
    pickle.dump(file, output)
    output.close()
    
filename = './shanghai_ershou.pkl' # 文件名
file = result_1_100 # 文件
pkl_save(filename, file)
```
